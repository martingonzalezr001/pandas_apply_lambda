```python
import pandas as pd
```

```python
#Leer csv en workshop
df = pd.read_csv("../data/input/IMDB-Movie-Data.csv")
df
```

```python
# We want to create bins of movies according to the number of votes they've received. For that matter, we will create a new column named 'bin' which will tag every movie as follow:

# From 0 to 999 ==> 'cat_1'
# From 1000 to 9999 ==> 'cat_2'
# From 10000 to 99999 ==> 'cat_3'
# From 100000 to 999999 ==> 'cat_4'
# More than 1000000 ==> 'cat_5'
df["Bin"] = df['Votes'].apply(lambda x: 'cat1' if x >= 0 and x <= 999 else
                              'cat2' if  x >= 1000 and x <= 9999 else
                              'cat3' if x >= 10000 and x <= 99999 else
                              'cat4' if x >= 100000 and x <= 999999 else
                              'cat5' if x >= 1000000 else
                              'undefined')
df
```

```python
# Challenge 2: We want to know how much is the revenue per minute for every movie.
def create_revenue_per_minute(revenue, minutes):
    revenue_per_minute = revenue / minutes
    return revenue_per_minute


```

```python
create_revenue_per_minute(333.13,121)
```

```python
df["Revenue_per_Minute"] = df.apply(lambda row: create_revenue_per_minute(row["Revenue (Millions)"], row["Runtime (Minutes)"]), axis=1)
```

```python
# Challenge 3 We want to create a new rating where we add 1 point if the genre is thriller but subtract 1 point if the genre is comedy.

def new_rating(genre, rating):
    new_rating = rating
    if genre == "Thriller":
        new_rating += 1
    if genre == "Comedy":
        new_rating -= 1
    return new_rating
    
    
```

```python
df["New_Rating"] = df.apply(lambda row: new_rating(row["Genre"], row["Rating"]), axis = 1)
```

```python
# Challenge 4: We want to know whether the integer part of the number resulting from the sum of the ASCII value of every character of the movie title divided by the number of votes, is a prime number (remember that prime numbers are integers).
# Nueva columna: bool, lo que devuelva el booleano de si el resultado de la operacion es un numero primo o no


#Funcion que calcula la suma de los valores de cada caracter en codigo ascii de los titulos de las peliculas
def ascii_number(title,votes):
    count = 0
    ascii_code = [ord(char) for char in title]
    for x in ascii_code:      
        count += x
    return count

```

```python
df['ASCII_sum']= df.apply(lambda row: ascii_number(row['Title'], row['Votes']), axis=1)
```

```python
def prime_number(ascii_number,votes):
    cociente = ascii_number / votes
    return cociente
```
